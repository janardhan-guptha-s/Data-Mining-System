```python
import pandas as pd
import numpy as np
data_heartdisease = pd.read_csv("heart_cleveland.csv", header = None)
data_heartdisease
```

```python
#Script implementing a series of steps
#1.Clean the dataset by removing any rows/columns with missing values. 
~data_heartdisease.isin(['?']).any(axis=1)
```

```python
data_heartdisease.isin(['?']).sum()
```

```python
#2. Randomly splitting the data into K equal folds by Setting K as 5
#importing required libraries
from sklearn.model_selection import StratifiedKFold
#Splitting K Equal Folds to 5
starkfold = StratifiedKFold(n_splits = 5)
js = data_heartdisease.drop(13, axis = 1)
g = data_heartdisease[13] 
js
```

```python
#importing required libraries
from sklearn.preprocessing import LabelEncoder
#Assigning Labels with Label Encoder
lbl = js. copy()
lbl_encoder = LabelEncoder()
for i in js.columns:
    lbl[i] = lbl_encoder.fit_transform(lbl[i]) 
js = lbl
js
```

```python
#3. Creating parameter grid which experimenting the gini and entropy impurity measures.
#importing required libraries
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
 
dtc_parameters = {
              'criterion' : ['gini', 'entropy'],
              'max_depth' : [10*len(data_heartdisease.columns)]    
             }
gini, entropy = [],[]
 
#Creating a loop for training a decision tree classifier for training and testing data
#Making sure maxi tree depth is set to avalue high enough for the dataset.
for train, test in starkfold.split(js,g):
    print("Training Data for DecisionTreeClassifier is:", train),
    print("Testing Data for DecisionTreeClassifier is:", test)
    js_train, js_test = js.loc[train,:], js.loc[test,:]
    g_train, g_test = g.loc[train], g.loc[test]
    for j in dtc_parameters['criterion']:
        decision_tree = DecisionTreeClassifier(criterion =j, max_depth = dtc_parameters['max_depth'][0])
        decision_tree.fit(js_train,g_train)
        dtc_scores = decision_tree.score(js_test,g_test)
        print('for {0} score is {1}'.format(j,dtc_scores))
        if j == 'gini':
            gini.append(dtc_scores)
        else:
            entropy.append(dtc_scores)
```

```python
#4.Computing overall accuracy for Gini 
accuracy_gini = sum(gini)/len(gini)
accuracy_gini
```

```python
#Computing the overall accuracy for Entropy.
accuracy_entropy = sum(entropy)/len(entropy)
accuracy_entropy
```
